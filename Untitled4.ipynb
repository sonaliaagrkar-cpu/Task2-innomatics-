```python
# Unique Words in a Sentence

# Taking input from user
sentence = input("Enter sentence: python is easy and python is powerful")


# Splitting sentence into words
words = sentence.split()

# Converting list to set to get unique words
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Highest Salary from Employee Data

# Given dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Printing result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
# Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

count = 0

# Loop to check each price
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

```python
# Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count number of presents
present_days = 0
for day in attendance:
    if day == "P":
        present_days += 1

# Calculate percentage
percentage = (present_days / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

```python
# Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Print result
print("Unique phone numbers:", unique_numbers)
```

```python

# Count Character Frequency

text = "python"

frequency = {}

# Loop through each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Print result
print(frequency)

```

```python
# Convert List to Tuple

numbers = [10, 20, 30]

# Convert list to tuple
numbers_tuple = tuple(numbers)

# Print result
print("Tuple:", numbers_tuple)
```

```python

# Check if a Key Exists in a Dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
